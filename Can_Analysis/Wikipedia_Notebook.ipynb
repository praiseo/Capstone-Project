```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import requests

import requests # library to handle requests


print('Libraries imported.')
```

```python

cannada_wikiPage = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')#Downloading data from Wikipedia
cannada_data=cannada_wikiPage.text
x=cannada_data.find("<table>")#getting column data here
y=cannada_data.find("</table>")
cannada_data=cannada_data[x+6:y]
x=cannada_data.find("<th>")
y=cannada_data.find("</th>")
col1=cannada_data[x+4:y]
cannada_data=cannada_data[y+5:]
x=cannada_data.find("<th>")
y=cannada_data.find("</th>")
col2=cannada_data[x+4:y]
cannada_data=cannada_data[y+4:]
x=cannada_data.find("<th>")
y=cannada_data.find("</th>")
col3=cannada_data[x+4:y-1]
cannada_data=cannada_data[y+2:]
columns_Data=[col1,col2,col3]
CanData=cannada_data#To make sure that cannada data doesn't get replaced and we don't need to run every module again and again
columns_Data
```

```python
cannada_data=CanData
CanList=[]
tdExists=True
while tdExists:
    i=0
    Item=[]
    while i<3:#finding table data
        x=cannada_data.find("<td>")
        y=cannada_data.find("</td>")
        if x==-1:
            tdExists=False
        v=cannada_data[x+4:y]
        if v.find("title") != -1:
            v=v[v.find("title=\""):]
            m=v.find(">")
            n=v.find("</a>")
            v=v[m+1:n]
        v=v.replace("\n","")
        Item.append(v)
        cannada_data=cannada_data[y+5:]
        i+=1
    if 'Not assigned' not in str(Item):#removing items not required
        CanList.append(Item)
del CanList[-1]

CanDat=pd.DataFrame(CanList, columns=columns_Data)
CanDat.shape#data shape
CanDat
```
